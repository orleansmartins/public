#Stock filter

São elegíveis para a carteira os seguintes ativos:

- Ações ordinárias;
- Ações preferenciais;
- Units;
- Todos os ativos precisam estar listados na B3;
- Aqueles com um mínimo de R$ 1 milhão de valor diário médio negociado.

```python
def is_ordinary(stocks):
    return pd.Series(stocks, index=stocks).str.endswith('3')

def is_preferencial(stocks):
    preferencial = pd.Series([False for _ in stocks], index=stocks)

    for code in ['4','5','6','7','8']:
        preferencial = preferencial | pd.Series(stocks, index=stocks).str.endswith(code)
    return preferencial

def is_unit(stocks):
    return pd.Series(stocks, index=stocks).str.endswith('11')
```

```python
def avg_volume_limit(volumes, limit=1e6):
    moving_avg = volumes.rolling(window=252).mean()
    return moving_avg > limit
```

```python
def has_positive_earnings(earnings):
    return earnings > 0
```

```python
def volatility_limit(prices, limit=0.5):
    volatility = prices.pct_change().rolling(window=21*11).std()*np.sqrt(11*21)
    return volatility < limit
```

```python
def evaluate_assets(prices, volumes, volume_inf_limit=1e7, volatility_sup_limit=0.5):
    assets_check = is_ordinary(volumes.columns) | is_preferencial(volumes.columns) | is_unit(volumes.columns)
    approved_tickers = assets_check[assets_check]

    approved_volumes = avg_volume_limit(volumes[approved_tickers.index.tolist()], volume_inf_limit)
    approved_volatilities = volatility_limit(prices, volatility_sup_limit)
    approved_earnings = has_positive_earnings(plucro)
    evaluation = (approved_volumes & approved_volatilities ).fillna(False)
    return evaluation
```

```python
def generate_portfolios(rolling_returns, approved_assets, n=5):
    dates = approved_assets.index

    long_cols = ['long_asset_'+str(i) for i in range(1, n+1)]
    short_cols = ['short_asset_'+str(i) for i in range(1, n+1)] 
    columns = long_cols + short_cols
    portfolios = pd.DataFrame(columns = columns, dtype=float)

    for d in range(1, len(dates)):
        date = dates[d]
        last_date = dates[d-1]
        i = rolling_returns.index.get_loc(last_date, method='pad')
        returns = rolling_returns.iloc[i]
        if len(returns.dropna()) <= 2*n:
            continue
        returns = returns.sort_values(ascending=False)
        assets = approved_assets.loc[date]
        returns = returns[assets]
        
        long_ports = returns[0:n].index.tolist() if len(returns) >= n else returns.index.tolist() + [np.nan for _ in range(len(returns), n)]

        if len(returns) >= 2*n:
            short_ports = returns[::-1][0:n].index.tolist()
        elif len(returns) < 2*n and len(returns) > n:
            short_ports = returns[n: n*2].index.tolist()+ [np.nan for _ in range(len(returns[n:2*n]), n)]
        else:
            short_ports = [np.nan for _ in range(n)]
        
        portfolios.loc[date] = long_ports + short_ports
   
    return portfolios
```

```python
portfolios = generate_portfolios(rolling_returns, approved_assets, n)
portfolios.index = portfolios.index + dt.timedelta(days=1)
long_portfolios = portfolios.drop(columns=[col for col in portfolios.columns if 'long' not in col])
short_portfolios = portfolios.drop(columns=[col for col in portfolios.columns if 'short' not in col])
```

```python
def get_portfolios_returns(portfolios, prices):
    portfolio_returns = pd.DataFrame(columns=['returns', 'long_returns', 'short_returns'], dtype=float)
    returns = prices.pct_change()
    long_portfolios = portfolios.drop(columns=[col for col in portfolios.columns if 'long' not in col])
    short_portfolios = portfolios.drop(columns=[col for col in portfolios.columns if 'short' not in col])

    for date in prices.index:
        if date < portfolios.index[0]:
            continue
        i = portfolios.index.get_loc(date, method='pad')
        long_p = long_portfolios.iloc[i].dropna().tolist()
        short_p = short_portfolios.iloc[i].dropna().tolist()

        long = returns.loc[date, long_p].mean()
        short = returns.loc[date, short_p].mean()
        if pd.isna(short):
            short = 0

        result = long - short

        portfolio_returns.loc[date] = [result, long, short]

    return portfolio_returns
```

```python
returns = get_portfolios_returns(portfolios, prices_studied)
acc_returns = ((returns+1).cumprod()-1)*100
returns *= 100
```
